```python
import csv
import pickle
from sklearn.neural_network import MLPClassifier

with open('NN_Classifier_700By3.Model', 'rb') as f:
    clf = pickle.load(f)
with open('testing_set.txt', 'rb') as f:
    X = pickle.load(f)
with open('testing_labels.txt', 'rb') as f:
    y = pickle.load(f)
with open('testing_Ids.txt', 'rb') as f:
    docIds = pickle.load(f)
    
p = clf.predict(X)
p_ = clf.predict_proba(X)
t_p = 0
t_n = 0
f_p = 0
f_n = 0

t_p_ids = []

for n in range(len(y)):
    if p[n] == y[n]:
        if p[n] == 1:
            t_p += 1;
           
        else:
           
            t_n += 1
    
    else:
        if p[n] == 0:
            t_p_ids.append(docIds[n])
            f_n += 1;
        else:
            f_p += 1;

if t_p+f_p == 0:
    precision = 0
else:
    precision = t_p / (t_p + f_p);
recall = t_p / (t_p + f_n);
if recall == 0:
    f_score = 0
else:
    f_score = (2 * precision * recall) / (precision + recall);
print("tp,fp,tn,fn: " + ', '.join([str(t_p), str(f_p), str(t_n), str(f_n)]))
print(precision);
print(recall);
print(f_score);
print("Accuracy: " + str( 100*(t_p+t_n)/len(p) ) )

with open('final_data_file.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        if row['docID'] in t_p_ids:
            print(row['text'])
            print('\n')
```
